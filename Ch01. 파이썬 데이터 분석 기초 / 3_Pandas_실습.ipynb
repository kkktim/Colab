```python
"""
날짜 : 2022/04/25
이름 : 강태호
내용 : 파이썬 Pandas 패키지 실습

Pandas 라이브러리
 - 파이썬 자료구조 DataFrame 처리를 지원하는 라이브러리
 - DataFrame은 행과 컬럼으로 이루어진 구조체(데이터베이스 테이블 구조)
"""
import pandas as pd
```

```python
#DataFrame 생성
df1 = pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]])
df1
```

```python
#칼럼명을 갖는 DataFrame 생성
df2 = pd.DataFrame({'uid': ['a101', 'a102', 'a103', 'a104', 'a105'],
                    'name': ['김춘추', '김유신','장보고','강감찬','이순신'],
                    'age': [25, 27, 32, 41, 52]})
df2
```

```python
#DataFrame 개발 출력
print(df1[0][0])
print(df1[0][1])
print(df1[1][2])
print(df2['uid'][0])
print(df2['name'][1])
print(df2['name'][4])
```
